```python
#Creating a backup copy
df_raw = df.copy()

#Start Cleaning on new dataframe
df_clean = df.copy()
```

```python
#Dropping unnecessary columns. unnamed is possibly a key doesnt contribute to analysis
df_clean.drop(columns=['Unnamed: 0', 'key'], inplace=True)
```

```python
#converting pickup_datetime to datetime
df_clean['pickup_datetime'] = pd.to_datetime(df_clean['pickup_datetime'], errors='coerce')
```

```python
#Filling missing values with column mean
df_clean['dropoff_longitude'] = df_clean['dropoff_longitude'].fillna(df_clean['dropoff_longitude'].mean())
df_clean['dropoff_latitude'] = df_clean['dropoff_latitude'].fillna(df_clean['dropoff_latitude'].mean())
```

```python
#Filtering out invalid fare values
df_clean = df_clean[(df_clean['fare_amount'] > 0) & (df_clean['fare_amount'] < 200)]
```

```python
#Filtering out invalid passenger counts
df_clean = df_clean[(df_clean['passenger_count'] >= 1) & (df_clean['passenger_count'] <= 6)]
```

```python
#Filtering unrealistic NYC Coordinates
df_clean = df_clean[
    (df_clean['pickup_latitude'].between(40.5, 41.0)) &
    (df_clean['pickup_longitude'].between(-74.5,-72.8)) &
    (df_clean['dropoff_latitude'].between(40.5, 41.0)) &
    (df_clean['dropoff_longitude'].between(-74.5,-72.8))
]
```

```python
df_clean.isnull().sum()
```

## **Feature Engineering**

```python
#Extracting features from datetime
df_clean['hour'] = df_clean['pickup_datetime'].dt.hour
df_clean['day'] = df_clean['pickup_datetime'].dt.day
df_clean['month'] = df_clean['pickup_datetime'].dt.month
df_clean['day_of_week'] = df_clean['pickup_datetime'].dt.day_name()
```

```python
# Flagging peak vs off-peak hours
def label_peak(hour):
    if 7 <= hour <= 9 or 17 <= hour <= 19:
        return 'peak'
    else:
        return 'Off-Peak'
    
df_clean['Peak_Offpeak'] = df_clean['hour'].apply(label_peak)
```
